```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Recycled_Commodities_and_Waste_Streams__Beginning_2008.csv")
```

```python
df.head()
```

```python
df_statewide = df[df["Region"] == "Statewide"]
df_region = df[df["Region"] != "Statewide"]
```

```python
df_statewide
```

```python
df_statewide.to_csv('df_statewide.csv', index='True')
```

# Region EDA

```python
df_region_cap = df_region[df_region["Region"] == "Region - Capital"]
df_region_down = df_region[df_region["Region"] == "Region - Downstate"]
df_region_up = df_region[df_region["Region"] == "Region - Upstate"]
```

```python
print(df_region_cap.info())
print(df_region_up.info())
print(df_region_down.info())

# all the same # of entries 
```

```python
df_region_cap
```

```python
region_wide = df_region_cap.pivot(index="Fiscal Year", columns="Type", values="Amount")
sns.lineplot(data=region_wide)
plt.title('Comparison between Recyled and Waste')
plt.ylabel('Amount in Tons')
plt.xlabel('Fiscal Year')
plt.show()
```

```python
region_upwide = df_region_up.pivot(index="Fiscal Year", columns="Type", values="Amount")
sns.lineplot(data=region_upwide)
plt.title('Comparison between Recyled and Waste for Upside')
plt.ylabel('Amount in Tons')
plt.xlabel('Fiscal Year')
plt.show()
```

```python
region_downwide = df_region_down.pivot(index="Fiscal Year", columns="Type", values="Amount")
sns.lineplot(data=region_downwide)
plt.title('Comparison between Recyled and Waste for Downside')
plt.ylabel('Amount in Tons')
plt.xlabel('Fiscal Year')
plt.show()
```

```python
df_region.to_csv('df_region.csv', index=True)
```

# State Wide EDA

```python
df_statewide["Type"].unique()
```

```python
df_st_per = df_statewide[(df_statewide["Type"] == 'Overall Waste Stream Reduction, Percentage') | (df_statewide["Type"] == 'Percentage of Waste Stream Recycled')]
```

```python
df_st_per = df_st_per.pivot(index='Fiscal Year', columns="Type", values="Amount")
```

```python
df_st_per
```

```python
sns.lineplot(data=df_st_per)
plt.title('Comparison between Recyled and Waste in % for StateWide')
plt.ylabel('Percentage')
plt.xlabel('Fiscal Year')
plt.show()
```

```python
df_st_per.to_csv("df_statewide_per.csv", index=True)
```

```python
df_st_ton = df_statewide[(df_statewide["Type"] == 'Total Tonnage of Recycled Material') | (df_statewide["Type"] == 'Total Tonnage of Waste Material')]
df_st_ton_piv = df_st_ton.pivot(index='Fiscal Year', columns="Type", values="Amount")
sns.lineplot(data=df_st_ton_piv)
plt.title('Comparison between Recyled and Waste in Tons for StateWide')
plt.ylabel('Tons')
plt.xlabel('Fiscal Year')
plt.show()
```

```python
df_st_ton
```

```python
df_st_ton_piv.to_csv("df_statewide_ton_piv.csv", index=True)
```
